```python
def parser(filename, path):
    data = set({})
    tags = set({})
    try:
        with zipfile.ZipFile(os.path.join(path, filename), 'r') as archive:
    
            namelist = archive.namelist()
            for archivename in namelist:
                if(archivename[-4:] != ".xml" and archivename[-5:] != ".rels" ): 
                    continue
                try:
                    with archive.open(archivename, mode = 'r') as f:

                        soup = BeautifulSoup(f, "lxml")
                        for tag in soup.find_all(re.compile("")):
                            tags.add(tag.name)
                        f.close()
                except:
                    print(os.path.join(path, filename, archivename))
            archive.close()
            
        data.update(tags)
        
    except:
        print(os.path.join(path, filename))
    return data
```

```python
def get_labels(filename):
    return filename[-1]
```

```python
def tags_to_array(df):
    name_tags = np.array(df)

    for value in name_tags:
        value[1] = value[1].replace("'","").replace("{","").replace("}","").split(", ")
    return name_tags


def get_unique_tags_train(np_data_train):
    unique_tags_set = set({})

    for value in np_data_train:
        unique_tags_set.update(set(value[1]))
    
    unique_tags = list(unique_tags_set)
    print(len(unique_tags))
    return unique_tags


def get_feature_tags(np_data, unique_tags):
    features = []
    for value in np_data:
        current_tags = np.zeros(len(unique_tags))
        for tag in value[1]:
            if tag in unique_tags:
                ind = unique_tags.index(tag)
                current_tags[ind] = 1
        features.append(current_tags)
        
    return features

def get_feature_error_exe(filename, path):
    
    error = 0
    exe = 0
    with open(os.path.join(os.path.join(path, filename)), mode = 'rb') as f:
        contents = f.read()
        try:
            zf = zipfile.ZipFile(io.BytesIO(contents), "r")
            for fileinfo in zf.infolist():
                try:
                    file = zf.read(fileinfo)
                    try:
                        decoded = file.decode("latin-1")
                        if ".exe" in decoded:
                            exe = 1
                    except:
                        error = 1
                        return [error, exe]
                except:
                    error = 1
                    return [error, exe]
                
        except:
            error = 1
            return [error, exe]
    return [error, exe]

def get_feature_errors_exe(filenames, path):
    
    errors = []
    exes = []
    for i, filename in enumerate(filenames):
        [error, exe] = get_feature_error_exe(filename, path)
        errors.append(error)
        exes.append(exe)
        
    return np.hstack([np.array(errors).astype(np.int).reshape(-1,1), np.array(exes).astype(np.int).reshape(-1,1)])
    
    
def get_all_features(filenames, path, np_data, unique_tags):
    
    features = get_feature_tags(np_data, unique_tags)
    
    errors_exes = get_feature_errors_exe(filenames, path)
    
    return np.hstack([np.array(features).astype(np.int), np.array(errors_exes).astype(np.int)])


def get_features_labels(filenames, path, np_data, unique_tags):

    all_features = get_all_features(filenames, path, np_data, unique_tags)
    
    labels = []
    for i, filename in enumerate(filenames):
        labels.append(get_labels(filename))
        
    data = np.hstack([np.array(all_features).astype(np.int), np.array(labels).astype(np.int).reshape(-1,1)])
    return data
```
