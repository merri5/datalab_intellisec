```python
import os
import pandas as pd
import numpy as np
from collections import OrderedDict

from joblib import Parallel, delayed

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV 
from sklearn.metrics import make_scorer, balanced_accuracy_score
```

```python
path_train = os.path.join(os.getcwd(), "ch02-train")
filenames_train = os.listdir(path_train)

path_test = os.path.join(os.getcwd(), "ch02-test")
filenames_test = os.listdir(path_test)
```

```python
def get_features(filename, path):
    dict_entry = {}
    file = [os.path.join(path, filename)]
    options = pdfid.get_fake_options()
    options.scan = True
    options.json = True
    
    error = 0
    features = np.array([])
    try:
        list_of_dict = pdfid.PDFiDMain(file, options)
        
        if list_of_dict["reports"] == []:
            error = 1
            features= np.append(np.zeros(21), error)
            dict_entry[filename] = features
            return dict_entry
        
    except:
        error = 1
        features= np.append(np.zeros(21),error)
        dict_entry[filename] = features
        return dict_entry
    features= np.append(features,list_of_dict["reports"][0]["obj"])
    features= np.append(features,list_of_dict["reports"][0]["endobj"])
    features= np.append(features,list_of_dict["reports"][0]["stream"])
    features= np.append(features,list_of_dict["reports"][0]["endstream"])
    features= np.append(features,list_of_dict["reports"][0]["xref"])
    features= np.append(features,list_of_dict["reports"][0]["trailer"])
    features= np.append(features,list_of_dict["reports"][0]["startxref"])
    features= np.append(features,list_of_dict["reports"][0]["/Page"])
    features= np.append(features,list_of_dict["reports"][0]["/Encrypt"])
    features= np.append(features,list_of_dict["reports"][0]["/ObjStm"])
    features= np.append(features,list_of_dict["reports"][0]["/JS"])
    features= np.append(features,list_of_dict["reports"][0]["/JavaScript"])
    features= np.append(features,list_of_dict["reports"][0]["/AA"])
    features= np.append(features,list_of_dict["reports"][0]["/OpenAction"])
    features= np.append(features,list_of_dict["reports"][0]["/AcroForm"])
    features= np.append(features,list_of_dict["reports"][0]["/JBIG2Decode"])
    features= np.append(features,list_of_dict["reports"][0]["/RichMedia"])
    features= np.append(features,list_of_dict["reports"][0]["/Launch"])
    features= np.append(features,list_of_dict["reports"][0]["/EmbeddedFile"])
    features= np.append(features,list_of_dict["reports"][0]["/XFA"])
    features= np.append(features,list_of_dict["reports"][0]["/Colors > 2^24"])
    features= np.append(features,error)
    dict_entry[filename] = features
    
    return dict_entry
```

# Train data

```python
%%time

data_train_list = (Parallel(n_jobs = -1)(delayed(get_features)(filename, path_train) for filename in filenames_train))
```

```python
data_train = {}

for entry in data_train_list:
    key = list(entry.keys())[0]
    data_train[key] = list(entry[key])
```

```python
dataframe_train = pd.DataFrame.from_dict(data_train, orient='index')
dataframe_train
```

```python
dataframe_train["y"] = pd.Series(dataframe_train.index).apply(lambda z: int(z[-1])).values
dataframe_train
```

```python
dataframe_train.to_csv("data_train_features.csv")
```

# Test data

```python
%%time

data_test_list = (Parallel(n_jobs = -1)(delayed(get_features)(filename, path_test) for filename in filenames_test))
```

```python
data_test = {}

for entry in data_test_list:
    key = list(entry.keys())[0]
    data_test[key] = list(entry[key])
```

```python
dataframe_test = pd.DataFrame.from_dict(data_test, orient='index')
```

```python
dataframe_test.to_csv("data_test_features.csv")
```

# Read saved data

```python
data_train1 = pd.read_csv("data_train_features.csv", index_col = 0)
data_train1
```

```python
train = data_train1.drop(["y", "20"], axis = 1).values
train.shape
```

```python
labels = data_train1.y.values
labels
```

```python
data_test1 = pd.read_csv("data_test_features.csv", index_col = 0)
data_test1
```

```python
test = data_test1.drop("20", axis = 1).values
test.shape
```

# Find parameters of the model

```python
# defining parameter range 
param_grid = {'n_estimators':  [50, 80, 90, 100, 110, 120, 150], 
              'class_weight':  ["None", "balanced"]}  

grid = GridSearchCV(RandomForestClassifier(), param_grid, refit = True, verbose = 3, n_jobs = -1, scoring= make_scorer(balanced_accuracy_score)) 
```

```python
grid.fit(train, labels) 
```

```python
grid.best_params_
```

# Model

```python
rf = RandomForestClassifier(class_weight= 'balanced', n_estimators = 100)
rf.fit(train, labels)

preds_rf = rf.predict(test)
```

```python
pd.Series(preds_rf, index = data_test1.index).to_csv("test_predictions.csv", header = None)
```
